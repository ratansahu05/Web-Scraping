# Web Scraping using API key

## Importing Libraries used to request data and pandas to store the data in the dataframe

```python
import pandas as pd
import requests 
```

```python
re  = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=f212f46a0ab5c3cd1fcb130038f4a461&language=en-US&page=1')
temp_df = pd.DataFrame(re.json()['results'])[['id', 'title', 'overview', 'vote_average', 'popularity']]
temp_df
```

```python
#Created empty dataframe to store data
df  = pd.DataFrame()
```

## Applied same request module to fetch the data from 100 pages by using for loop

```python
for i in range(1, 101):
    re  = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=f212f46a0ab5c3cd1fcb130038f4a461&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(re.json()['results'])[['id', 'title', 'overview', 'vote_average', 'popularity']]
    df = pd.concat([df, temp_df], ignore_index = True)
    
df
```
